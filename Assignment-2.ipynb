# Feature Engineering

```python
# Get difference between visitor and hotel rating
train_data["diff_starrating"] = train_data["visitor_hist_starrating"] - train_data["prop_starrating"]
test_data["diff_starrating"] = test_data["visitor_hist_starrating"] - test_data["prop_starrating"]
```

```python
train_data["price_usd_normalized"] = train_data["price_usd"] / train_data["srch_length_of_stay"]
test_data["price_usd_normalized"] = test_data["price_usd"] / test_data["srch_length_of_stay"]
```

```python
# standardize 'srch_adults_count', 'srch_children_count', 'srch_room_count'
ad_mean=train_data['srch_adults_count'].mean()
ad_std=train_data['srch_adults_count'].std()
ch_mean=train_data['srch_children_count'].mean()
ch_std=train_data['srch_children_count'].std()

room_mean=train_data['srch_room_count'].mean()
room_std=train_data['srch_room_count'].std()

train_data['srch_adults_count']=(train_data['srch_adults_count']-ad_mean)/ad_std
train_data['srch_children_count']=(train_data['srch_children_count']-ch_mean)/ch_std
train_data['srch_room_count']=(train_data['srch_room_count']-room_mean)/room_std
```

```python
ad_mean=test_data['srch_adults_count'].mean()
ad_std=test_data['srch_adults_count'].std()
ch_mean=test_data['srch_children_count'].mean()
ch_std=test_data['srch_children_count'].std()

room_mean=test_data['srch_room_count'].mean()
room_std=test_data['srch_room_count'].std()

test_data['srch_adults_count']=(test_data['srch_adults_count']-ad_mean)/ad_std
test_data['srch_children_count']=(test_data['srch_children_count']-ch_mean)/ch_std
test_data['srch_room_count']=(test_data['srch_room_count']-room_mean)/room_std
```

```python
#Adding Score columns: 5 for booked, 1 clicked and 0 the rest
def assign_score(x):
    if x["booking_bool"]==1:
        val=5
    elif x["click_bool"]==1:
        val=1
    else:
        val=0
    return val


train_data['score'] = train_data.apply(assign_score , axis=1)
```

```python
train_data = train_data.drop(columns= ['click_bool', 'booking_bool'], axis=1)
```

### KNN - GridSearch

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, ndcg_score


features = [
    'visitor_hist_starrating', 'prop_starrating', 'prop_review_score',
    'prop_location_score1', 'prop_location_score2',
    'srch_length_of_stay', 'srch_booking_window'
]
target = 'booking_bool'


X = final_train[features].fillna(final_train[features].median())  # Fill missing values
y = final_train[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = KNeighborsClassifier()
param_grid = {
    'n_neighbors': [3, 5, 7, 10],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

def ndcg_scorer(y, y_pred, **kwargs):
    y_true = [[1 if x == y.iloc[i] else 0 for x in np.unique(y)] for i in range(len(y))]
    return ndcg_score(y_true, y_pred, **kwargs)

scorer = make_scorer(ndcg_scorer, needs_proba=True, k=5)

grid_search = GridSearchCV(model, param_grid, cv=5, scoring=scorer)
grid_search.fit(X_train, y_train)


print("Best parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_

y_pred_proba = best_model.predict_proba(X_test)
ndcg_value = ndcg_scorer(y_test, y_pred_proba, k=5)
print("NDCG@5 Score on test set:", ndcg_value)

X_test_final = test_data[features].fillna(test_data[features].median())
X_test_final_scaled = scaler.transform(X_test_final)
test_data['booking_probability'] = best_model.predict_proba(X_test_final_scaled)[:, 1]

sorted_test_data = test_data.sort_values(['srch_id', 'booking_probability'], ascending=[True, False])

submission = sorted_test_data[['srch_id', 'prop_id']]
submission.to_csv('submission.csv', index=False, header=True)
```
